Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful data visualization library in Python that provides a variety of plot types to visualize different aspects of your data. Here are five common plots you can create using Seaborn, along with their primary uses:

Scatter Plot (scatterplot()): A scatter plot is used to visualize the relationship between two continuous variables. It displays individual data points as markers on a Cartesian coordinate system, allowing you to observe patterns, clusters, or correlations between variables.

Bar Plot (barplot()): A bar plot is used to compare categorical variables or display the distribution of a categorical variable. It represents the data using rectangular bars, where the height or length of each bar corresponds to the value being plotted. Bar plots are useful for visualizing frequencies, proportions, or summary statistics across different categories.

Histogram (histplot()): A histogram is used to represent the distribution of a single continuous variable. It displays the frequencies or counts of data points within specified intervals (bins) along the x-axis, allowing you to visualize the shape, central tendency, and spread of the data.

Box Plot (boxplot()): A box plot (or box-and-whisker plot) is used to summarize the distribution of a continuous variable across different categories. It displays the median, quartiles, and any outliers in the data, providing insights into the central tendency, spread, and skewness of the distributions.

Heatmap (heatmap()): A heatmap is used to visualize the relationship between two categorical variables or display the correlation matrix of a dataset. It represents the data using a grid of colored cells, where the color intensity or shade indicates the strength or magnitude of the relationship or correlation. Heatmaps are effective for identifying patterns, clusters, or associations in large datasets.

These are just a few examples of the plot types available in Seaborn. Seaborn also provides other plot types, such as line plots (lineplot()), violin plots (violinplot()), and many more, allowing you to choose the appropriate visualization technique based on your data and analytical goals.

4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
sns.load_dataset("diamonds")
```

```python

import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds, x='price', hue='cut', kde=True)

plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Price by Cut')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_df = sns.load_dataset("flights")

# Reshape the dataset for heatmap
flights_pivot = flights_df.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Set the title and labels
plt.title("Passenger Numbers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the heatmap
plt.show()
```
